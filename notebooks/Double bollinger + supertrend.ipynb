```python
import yfinance as yf
import pandas_ta as ta
import nsepy as nse
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
from datetime import date
import vectorbt as vbt
import backtrader as bt
from dateutil.relativedelta import relativedelta
%matplotlib inline
```

```python
nsei = yf.Ticker('^NSEI')
start_date = date(2017,8,31)
end_date = date.today()

nsei_df = nsei.history(start=start_date,
                   end=end_date)

ax = nsei_df[['Close']].plot(figsize=(20,10))
```

```python
bbands1 = ta.bbands(nsei_df['Close'], length=20, std=1)
bbands2 = ta.bbands(nsei_df['Close'], length=20, std=2)

nsei_df[['Upper1', 'Lower1', 'Middle1']] = bbands1[['BBU_20_1.0', 'BBL_20_1.0', 'BBM_20_1.0']]
nsei_df[['Upper2', 'Lower2', 'Middle2']] = bbands2[['BBU_20_2.0', 'BBL_20_2.0', 'BBM_20_2.0']]
nsei_df
```
